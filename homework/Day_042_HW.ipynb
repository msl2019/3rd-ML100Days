```python
from sklearn import datasets, metrics

# 如果是分類問題，請使用 DecisionTreeClassifier，若為回歸問題，請使用 DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
# Wine
#嘗試使用不同的方法(gini,entropy)
#和 min_samples_split,min_samples_leaf, 
wine = datasets.load_wine()
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=4)
x_method = ['gini','entropy']
x_split =[2,3,4]
x_leaf = [1,2,3]
```

```python
for b_method in x_method:
    for b_split in x_split:
        for b_leaf in x_leaf:
            clf = DecisionTreeClassifier(
                criterion = b_method,
                max_depth = None,
                min_samples_split = b_split,
                min_samples_leaf =b_leaf                
            )
            clf.fit(x_train, y_train)
            y_pred = clf.predict(x_test)
            acc = metrics.accuracy_score(y_test, y_pred)
            print("-----------------------------------------------------------")
            print("Acuuracy: ", acc)
            print(wine.feature_names)
            print("Feature importance: ", clf.feature_importances_)
            print("Method:",b_method, ",Min_sample_Split:",b_split,",min_samples_leaf:",b_leaf,"\n")
```
