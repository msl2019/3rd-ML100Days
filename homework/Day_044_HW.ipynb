```python
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# Wine
#嘗試使用不同的方法(gini,entropy), Estimators
#和 min_samples_split,min_samples_leaf,depth

wine = datasets.load_wine()
x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.25, random_state=4)
x_method = ['gini','entropy']
x_estimators =[10,20,30]
x_depth = [2,5,10]
x_split =[2,3,4]
x_leaf = [1,2,3]
```

```python
for b_method in x_method:
    for b_estimators in x_estimators:
        for b_depth in x_depth:
            for b_split in x_split:
                for b_leaf in x_leaf:
                    clf = RandomForestClassifier(
                        n_estimators=b_estimators, 
                        criterion=b_method,
                        max_features="auto",
                        max_depth=b_depth,
                        min_samples_split=b_split,
                        min_samples_leaf=b_leaf
                    )

                    clf.fit(x_train, y_train)
                    y_pred = clf.predict(x_test)
                    acc = metrics.accuracy_score(y_test, y_pred)
                    print("Accuracy: ", acc)
                    print(wine.feature_names)
                    print("Feature importance: ", clf.feature_importances_)
                    print("Estimators:",b_estimators, ",Method:",b_method, ",max_depth:",b_depth, ",Min_sample_Split:",b_split,",min_samples_leaf:",b_leaf,"\n")
                    print("-----------------------------------------------------------")

                    
```
