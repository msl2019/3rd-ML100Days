```python
#計算任意欄位的平均數
print ("AMT_ANNUITY mean value:",app_train.loc[:,"AMT_ANNUITY"].mean())
```

```python
#計算任意欄位的平均數
print ("AMT_GOODS_PRICE:",app_train.loc[:,"AMT_GOODS_PRICE"].mean())
```

```python
#計算任意欄位的標準差
print ("AMT_ANNUITY standard deviation value:",app_train.loc[:,"AMT_ANNUITY"].std())
```

```python
#計算任意欄位的標準差
print ("AMT_GOODS_PRICE standard deviation value:",app_train.loc[:,"AMT_GOODS_PRICE"].std())
```

```python
#畫出任意欄位的直方圖
# AMT_ANNUITY

plt.figure(figsize=[10,8])
app_train['AMT_ANNUITY'].hist(bins=50, range = (0, 140000), color = '#0504aa')
plt.title('AMT ANNUITY')
plt.xlabel("AMT_ANNUITY")
plt.ylabel("Frequency")
plt.show()
```

```python
#畫出任意欄位的直方圖
# AMT_GOODS_PRICE

plt.figure(figsize=[10,8])
app_train['AMT_GOODS_PRICE'].hist(bins=50, range = (0, 2500000), color = '#3a3ccc')
plt.title('AMT GOODS PRICE')
plt.xlabel("AMT_GOODS_PRICE")
plt.ylabel("Frequency")
plt.show()
```
