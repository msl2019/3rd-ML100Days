```python
from sklearn import datasets, metrics
digits = datasets.load_digits()
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
# 切分訓練集/測試集 Try different learning rate
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=4)
x_learning = [0.1,0.05,1,0.01]
x_max_leaf_notes = [2,3,4,5]
x_estimators = [50,100,200]
```

```python
for b_learning in x_learning:
    for b_max in x_max_leaf_notes: 
        for b_estimators in x_estimators:            
            clf = GradientBoostingClassifier(
                loss="deviance",
                learning_rate = b_learning,
                n_estimators= b_estimators,
                max_leaf_nodes = b_max
                )
            clf.fit(x_train, y_train)
            y_pred = clf.predict(x_test)
            acc = metrics.accuracy_score(y_test, y_pred)
            print("Acuuracy: ", acc)
            print("learning_rate:",b_learning,",max_leaf_nodes:" ,b_max, " ,n_estimators:",b_estimators)
            print("-------------------------")
```
